## 📊 3. REE Dataset Construction

Build comprehensive REE patent dataset using combined keyword and CPC classification search strategies optimized for the 2010-2023 timeframe.

```python
# Import dataset builder
from dataset_builder import build_ree_dataset, enrich_dataset_with_titles_abstracts, validate_dataset_structure

# Configure analysis mode
# Set to False for full production analysis (may take longer)
TEST_MODE = True

print("📊 Building REE Patent Dataset...")
print(f"Mode: {'TEST (Limited)' if TEST_MODE else 'PRODUCTION (Full)'}")
print("\n🔍 Search Strategy:")
print("   • Keywords: rare earth elements, neodymium, dysprosium, yttrium, lanthanide, REE recycling")
print("   • CPC Classifications: C22B% (Metallurgy), Y02W30% (Recycling), H01F1% (Magnets)")
print("   • Timeframe: 2010-2023 (14 years comprehensive coverage)")

# Build dataset
ree_dataset = build_ree_dataset(db, test_mode=TEST_MODE)

if not ree_dataset.empty:
    # Enrich with full text data
    print("\n📄 Enriching with title and abstract data...")
    enriched_dataset = enrich_dataset_with_titles_abstracts(db, ree_dataset)
    
    # Validate structure
    if validate_dataset_structure(enriched_dataset):
        print("\n✅ REE Dataset Successfully Created")
        print(f"   📊 Total Applications: {len(enriched_dataset):,}")
        print(f"   📁 Unique Families: {enriched_dataset['docdb_family_id'].nunique():,}")
        print(f"   📅 Year Range: {enriched_dataset['appln_filing_year'].min()}-{enriched_dataset['appln_filing_year'].max()}")
        print(f"   🌍 Filing Authorities: {enriched_dataset['appln_auth'].nunique()}")
        
        # Display sample data
        print("\n📋 Sample Dataset (First 3 Records):")
        display(enriched_dataset[['appln_id', 'appln_filing_year', 'appln_auth', 'appln_title']].head(3))
    else:
        raise Exception("Dataset validation failed")
else:
    raise Exception("No REE dataset created - check search parameters")
```

## 🔗 4. Citation Intelligence Analysis

Perform comprehensive forward and backward citation analysis using proper PATSTAT publication linkage methodology.

```python
# Import citation analysis modules
from citation_analyzer import (
    get_forward_citations, 
    get_backward_citations, 
    analyze_citation_patterns, 
    calculate_citation_metrics
)

# Extract application IDs for citation analysis
appln_ids = enriched_dataset['appln_id'].tolist()

print("🔗 CITATION INTELLIGENCE ANALYSIS")
print(f"Analyzing citations for {len(appln_ids):,} REE applications...")
print("\n📈 Forward Citations (Who cites our REE patents):")

# Forward citation analysis
forward_citations = get_forward_citations(db, appln_ids, test_mode=TEST_MODE)

print("\n📉 Backward Citations (What our REE patents cite):")

# Backward citation analysis  
backward_citations = get_backward_citations(db, appln_ids, test_mode=TEST_MODE)

print("\n🧠 Citation Pattern Analysis:")

# Analyze citation patterns
citation_patterns = analyze_citation_patterns(forward_citations, backward_citations)

print("\n📊 Citation Metrics Calculation:")

# Calculate citation metrics per application
citation_metrics = calculate_citation_metrics(enriched_dataset, forward_citations, backward_citations)

# Display citation summary
print("\n✅ CITATION ANALYSIS COMPLETE")
print(f"   🔗 Total Citation Network: {len(forward_citations) + len(backward_citations):,} citations")
print(f"   📈 Forward Citations: {len(forward_citations):,}")
print(f"   📉 Backward Citations: {len(backward_citations):,}")
print(f"   🎯 Citation Balance: {citation_patterns.get('citation_balance', 'N/A')}")

if not citation_metrics.empty:
    high_impact = (citation_metrics['impact_category'] == 'High Impact').sum()
    print(f"   ⭐ High Impact Applications: {high_impact}")
    
    # Display top cited applications
    if 'forward_citations' in citation_metrics.columns:
        top_cited = citation_metrics.nlargest(3, 'forward_citations')[['appln_id', 'appln_filing_year', 'forward_citations', 'impact_category']]
        print("\n🏆 Top Cited REE Applications:")
        display(top_cited)
```

```python
# Create business intelligence visualizations
print("📊 BUSINESS INTELLIGENCE VISUALIZATIONS")
print("Creating executive-ready charts and dashboards...")

# Set up the plotting environment
fig, axes = plt.subplots(2, 2, figsize=(16, 12))
fig.suptitle('REE Patent Intelligence Dashboard - Executive Summary', fontsize=16, fontweight='bold')

# 1. Patent Filing Trends Over Time
if 'appln_filing_year' in geo_enriched_data.columns:
    yearly_counts = geo_enriched_data['appln_filing_year'].value_counts().sort_index()
    axes[0, 0].plot(yearly_counts.index, yearly_counts.values, marker='o', linewidth=2, markersize=6)
    axes[0, 0].set_title('REE Patent Filing Trends (2010-2023)', fontweight='bold')
    axes[0, 0].set_xlabel('Filing Year')
    axes[0, 0].set_ylabel('Number of Applications')
    axes[0, 0].grid(True, alpha=0.3)

# 2. Top Filing Authorities
if 'appln_auth' in geo_enriched_data.columns:
    top_authorities = geo_enriched_data['appln_auth'].value_counts().head(8)
    axes[0, 1].bar(top_authorities.index, top_authorities.values, color='lightcoral')
    axes[0, 1].set_title('Top Patent Filing Authorities', fontweight='bold')
    axes[0, 1].set_xlabel('Filing Authority')
    axes[0, 1].set_ylabel('Number of Applications')
    axes[0, 1].tick_params(axis='x', rotation=45)

# 3. Citation Impact Distribution
if not citation_metrics.empty and 'impact_category' in citation_metrics.columns:
    impact_dist = citation_metrics['impact_category'].value_counts()
    colors = ['#ff6b6b', '#4ecdc4', '#45b7d1', '#96ceb4']
    axes[1, 0].pie(impact_dist.values, labels=impact_dist.index, autopct='%1.1f%%', 
                   colors=colors, startangle=90)
    axes[1, 0].set_title('Citation Impact Distribution', fontweight='bold')

# 4. International Collaboration Overview
if collaboration:
    collab_data = {
        'Single Country\nApplicants': collaboration.get('single_country_applicants', 0),
        'Multi Country\nApplicants': collaboration.get('multi_country_applicants', 0),
        'Single Country\nInventors': collaboration.get('single_country_inventors', 0),
        'Multi Country\nInventors': collaboration.get('multi_country_inventors', 0)
    }
    
    categories = list(collab_data.keys())
    values = list(collab_data.values())
    
    axes[1, 1].bar(categories, values, color=['lightblue', 'darkblue', 'lightgreen', 'darkgreen'])
    axes[1, 1].set_title('International Collaboration Patterns', fontweight='bold')
    axes[1, 1].set_ylabel('Number of Applications')
    axes[1, 1].tick_params(axis='x', rotation=45)

plt.tight_layout()
plt.subplots_adjust(top=0.93)
plt.show()

print("✅ Business intelligence visualizations complete")
print("📈 Executive dashboard ready for stakeholder presentations")
```

## 💾 8. Professional Data Exports

Export analysis results in business-ready formats for further analysis and reporting.

```python
# Export analysis results for business use
print("💾 PROFESSIONAL DATA EXPORTS")
print("Creating business-ready data files...")

export_timestamp = datetime.now().strftime('%Y%m%d_%H%M')
export_prefix = f"REE_Analysis_{export_timestamp}"

exported_files = []

try:
    # 1. Main REE dataset with geographic enrichment
    if not geo_enriched_data.empty:
        main_file = f"{export_prefix}_REE_Dataset.csv"
        geo_enriched_data.to_csv(main_file, index=False)
        exported_files.append(main_file)
        print(f"✅ Main Dataset: {main_file} ({len(geo_enriched_data):,} records)")
    
    # 2. Citation metrics for business intelligence
    if not citation_metrics.empty:
        metrics_file = f"{export_prefix}_Citation_Metrics.csv"
        citation_metrics.to_csv(metrics_file, index=False)
        exported_files.append(metrics_file)
        print(f"✅ Citation Metrics: {metrics_file} ({len(citation_metrics):,} records)")
    
    # 3. Forward citations network
    if not forward_citations.empty:
        forward_file = f"{export_prefix}_Forward_Citations.csv"
        forward_citations.to_csv(forward_file, index=False)
        exported_files.append(forward_file)
        print(f"✅ Forward Citations: {forward_file} ({len(forward_citations):,} records)")
    
    # 4. Backward citations network
    if not backward_citations.empty:
        backward_file = f"{export_prefix}_Backward_Citations.csv"
        backward_citations.to_csv(backward_file, index=False)
        exported_files.append(backward_file)
        print(f"✅ Backward Citations: {backward_file} ({len(backward_citations):,} records)")
    
    # 5. Geographic analysis summary (JSON)
    if geographic_analysis:
        geo_file = f"{export_prefix}_Geographic_Analysis.json"
        with open(geo_file, 'w') as f:
            json.dump(geographic_analysis, f, indent=2, default=str)
        exported_files.append(geo_file)
        print(f"✅ Geographic Analysis: {geo_file}")
    
    # 6. Quality report (JSON)
    if quality_metrics:
        quality_file = f"{export_prefix}_Quality_Report.json"
        with open(quality_file, 'w') as f:
            json.dump(quality_metrics, f, indent=2, default=str)
        exported_files.append(quality_file)
        print(f"✅ Quality Report: {quality_file}")
    
    # 7. Executive summary (simplified CSV)
    executive_summary = {
        'Metric': [
            'Total REE Applications',
            'Unique Patent Families', 
            'Filing Authorities',
            'Applicant Countries',
            'Forward Citations',
            'Backward Citations',
            'Quality Score',
            'Business Readiness'
        ],
        'Value': [
            len(geo_enriched_data),
            geo_enriched_data['docdb_family_id'].nunique() if 'docdb_family_id' in geo_enriched_data.columns else 0,
            geo_enriched_data['appln_auth'].nunique() if 'appln_auth' in geo_enriched_data.columns else 0,
            len(geographic_analysis.get('applicant_countries', {})),
            len(forward_citations),
            len(backward_citations),
            f"{overall_score}/100",
            readiness_level
        ]
    }
    
    exec_file = f"{export_prefix}_Executive_Summary.csv"
    pd.DataFrame(executive_summary).to_csv(exec_file, index=False)
    exported_files.append(exec_file)
    print(f"✅ Executive Summary: {exec_file}")
    
    print(f"\n🎯 EXPORT COMPLETE")
    print(f"   📁 Files Created: {len(exported_files)}")
    
    # Calculate total export size
    total_size = sum(os.path.getsize(f) for f in exported_files if os.path.exists(f))
    size_mb = total_size / (1024 * 1024)
    print(f"   💾 Total Size: {size_mb:.1f} MB")
    print(f"\n 📊 Ready for stakeholder distribution and further analysis")
    
except Exception as e:
    print(f"❌ Export error: {e}")

# Display exported files list
if exported_files:
    print("\n📋 Exported Files:")
    for i, file in enumerate(exported_files, 1):
        print(f"   {i}. {file}")
```
