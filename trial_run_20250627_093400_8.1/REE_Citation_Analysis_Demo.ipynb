## üìà Step 4: Innovation Trends & Technology Evolution

**Business Value**: Technology lifecycle analysis showing innovation peaks, emerging trends, and market maturity indicators.

```python
if results and not ree_dataset.empty:
    print("üìà INNOVATION TRENDS ANALYSIS")
    print("=" * 35)
    
    # Yearly filing trends
    yearly_filings = ree_dataset['appln_filing_year'].value_counts().sort_index()
    
    print(f"\nüìä REE Patent Activity (2010-2023):")
    for year, count in yearly_filings.items():
        if year >= 2010:  # Focus on our analysis period
            print(f"   {year}: {count:,} applications")
    
    # Innovation trend visualization
    plt.figure(figsize=(14, 8))
    
    # Subplot 1: Filing trends over time
    plt.subplot(2, 2, 1)
    recent_years = yearly_filings[yearly_filings.index >= 2010]
    plt.plot(recent_years.index, recent_years.values, marker='o', linewidth=2, markersize=6)
    plt.title('REE Patent Filings Over Time')
    plt.xlabel('Year')
    plt.ylabel('Number of Applications')
    plt.grid(True, alpha=0.3)
    
    # Subplot 2: Search method effectiveness
    plt.subplot(2, 2, 2)
    if 'search_method' in ree_dataset.columns:
        search_dist = ree_dataset['search_method'].value_counts()
        plt.bar(range(len(search_dist)), search_dist.values)
        plt.xticks(range(len(search_dist)), search_dist.index, rotation=45)
        plt.title('Search Strategy Effectiveness')
        plt.ylabel('Patents Found')
    
    # Subplot 3: Technology maturity indicators
    plt.subplot(2, 2, 3)
    if citation_results:
        forward_cit = citation_results.get('forward_citations', pd.DataFrame())
        if not forward_cit.empty and 'citing_year' in forward_cit.columns:
            citation_trends = forward_cit['citing_year'].value_counts().sort_index()
            recent_citations = citation_trends[citation_trends.index >= 2010]
            plt.plot(recent_citations.index, recent_citations.values, 
                    marker='s', color='red', linewidth=2, markersize=6)
            plt.title('REE Technology Citations Over Time')
            plt.xlabel('Year')
            plt.ylabel('Citations Received')
            plt.grid(True, alpha=0.3)
    
    # Subplot 4: Market concentration
    plt.subplot(2, 2, 4)
    if 'primary_applicant_country' in ree_dataset.columns:
        top_6_countries = ree_dataset['primary_applicant_country'].value_counts().head(6)
        plt.barh(range(len(top_6_countries)), top_6_countries.values)
        plt.yticks(range(len(top_6_countries)), top_6_countries.index)
        plt.title('Market Leaders (Top 6 Countries)')
        plt.xlabel('Number of Patents')
    
    plt.tight_layout()
    plt.show()
    
    # Technology lifecycle assessment
    total_apps = len(ree_dataset)
    recent_growth = yearly_filings.tail(3).mean() / yearly_filings.head(3).mean() if len(yearly_filings) >= 6 else 1
    
    print(f"\nüî¨ Technology Lifecycle Assessment:")
    if total_apps >= 1000:
        print(f"   ‚Ä¢ Market Status: Mature technology domain ({total_apps:,} patents)")
    elif total_apps >= 300:
        print(f"   ‚Ä¢ Market Status: Developing technology ({total_apps:,} patents)")
    else:
        print(f"   ‚Ä¢ Market Status: Emerging technology ({total_apps:,} patents)")
    
    if recent_growth > 1.2:
        print(f"   ‚Ä¢ Growth Pattern: Accelerating innovation ({recent_growth:.1f}x growth)")
    elif recent_growth > 0.8:
        print(f"   ‚Ä¢ Growth Pattern: Stable innovation activity")
    else:
        print(f"   ‚Ä¢ Growth Pattern: Declining activity (mature/saturated market)")
```

## üîó Step 5: Citation Network Intelligence

**Business Value**: Innovation impact assessment, technology transfer patterns, and prior art landscape for strategic decision-making.

```python
if results and citation_results:
    print("üîó CITATION NETWORK INTELLIGENCE")
    print("=" * 40)
    
    forward_cit = citation_results.get('forward_citations', pd.DataFrame())
    backward_cit = citation_results.get('backward_citations', pd.DataFrame())
    
    print(f"\nüìä Citation Network Overview:")
    print(f"   ‚Ä¢ Forward Citations: {len(forward_cit):,} (impact on future innovation)")
    print(f"   ‚Ä¢ Backward Citations: {len(backward_cit):,} (building on prior art)")
    
    if len(forward_cit) > 0 and len(backward_cit) > 0:
        innovation_ratio = len(forward_cit) / len(backward_cit)
        print(f"   ‚Ä¢ Innovation Ratio: {innovation_ratio:.2f} (forward/backward)")
        
        if innovation_ratio > 1.2:
            print(f"   ‚Üí High-impact technology: Strong influence on future innovations")
        elif innovation_ratio > 0.8:
            print(f"   ‚Üí Balanced innovation: Good foundation + future potential")
        else:
            print(f"   ‚Üí Knowledge-building: Strong foundation, emerging impact")
    
    # Citation origin analysis
    citation_origins_ref = {
        'SEA': 'Search Report (Examiner)',
        'APP': 'Applicant Self-Citation',
        'ISR': 'International Search',
        'EXA': 'Examiner Citation',
        'OPP': 'Opposition Proceeding'
    }
    
    if not forward_cit.empty and 'citn_origin' in forward_cit.columns:
        print(f"\nüìã Forward Citation Sources:")
        origin_counts = forward_cit['citn_origin'].value_counts()
        for origin, count in origin_counts.items():
            origin_name = citation_origins_ref.get(origin, origin)
            percentage = (count / len(forward_cit)) * 100
            print(f"   ‚Ä¢ {origin_name}: {count:,} ({percentage:.1f}%)")
    
    # Geographic citation patterns
    if not forward_cit.empty and 'citing_country' in forward_cit.columns:
        citing_countries = forward_cit['citing_country'].value_counts().head(5)
        print(f"\nüåç Top Countries Citing REE Technology:")
        for i, (country, count) in enumerate(citing_countries.items(), 1):
            percentage = (count / len(forward_cit)) * 100
            print(f"   {i}. {country}: {count:,} citations ({percentage:.1f}%)")
    
    # Citation network visualization
    if not forward_cit.empty or not backward_cit.empty:
        plt.figure(figsize=(15, 10))
        
        # Citation trends over time
        plt.subplot(2, 3, 1)
        if not forward_cit.empty and 'citing_year' in forward_cit.columns:
            forward_trends = forward_cit['citing_year'].value_counts().sort_index()
            recent_forward = forward_trends[forward_trends.index >= 2010]
            plt.plot(recent_forward.index, recent_forward.values, 
                    marker='o', color='green', label='Forward Citations')
        
        if not backward_cit.empty and 'cited_year' in backward_cit.columns:
            backward_trends = backward_cit['cited_year'].dropna().astype(int).value_counts().sort_index()
            recent_backward = backward_trends[backward_trends.index >= 2000]
            plt.plot(recent_backward.index, recent_backward.values, 
                    marker='s', color='blue', alpha=0.7, label='Backward Citations')
        
        plt.title('Citation Patterns Over Time')
        plt.xlabel('Year')
        plt.ylabel('Number of Citations')
        plt.legend()
        plt.grid(True, alpha=0.3)
        
        # Citation origin distribution
        if not forward_cit.empty and 'citn_origin' in forward_cit.columns:
            plt.subplot(2, 3, 2)
            origin_dist = forward_cit['citn_origin'].value_counts().head(6)
            plt.pie(origin_dist.values, labels=origin_dist.index, autopct='%1.1f%%')
            plt.title('Forward Citation Origins')
        
        # Geographic citation distribution
        if not forward_cit.empty and 'citing_country' in forward_cit.columns:
            plt.subplot(2, 3, 3)
            country_dist = forward_cit['citing_country'].value_counts().head(8)
            plt.barh(range(len(country_dist)), country_dist.values)
            plt.yticks(range(len(country_dist)), country_dist.index)
            plt.title('Countries Citing REE Technology')
            plt.xlabel('Number of Citations')
        
        # Technology transfer matrix
        if not ree_dataset.empty and not forward_cit.empty:
            plt.subplot(2, 3, 4)
            # Create a simplified technology transfer visualization
            ree_countries = ree_dataset['appln_auth'].value_counts().head(5)
            citing_countries = forward_cit['citing_country'].value_counts().head(5)
            
            plt.bar(range(len(ree_countries)), ree_countries.values, 
                   alpha=0.7, label='REE Patent Origins')
            plt.bar(range(len(citing_countries)), citing_countries.values, 
                   alpha=0.7, label='Citation Sources')
            plt.title('Technology Transfer Patterns')
            plt.legend()
        
        plt.tight_layout()
        plt.show()
    
    # Strategic citation insights
    print(f"\nüéØ Strategic Citation Intelligence:")
    
    if len(forward_cit) > 100:
        print(f"   ‚Ä¢ High Citation Impact: REE technology actively influencing innovation")
    elif len(forward_cit) > 20:
        print(f"   ‚Ä¢ Moderate Citation Impact: Growing technology influence")
    else:
        print(f"   ‚Ä¢ Emerging Citation Impact: Early-stage technology development")
    
    if len(backward_cit) > len(forward_cit) * 2:
        print(f"   ‚Ä¢ Innovation Pattern: Building heavily on existing knowledge base")
    elif len(forward_cit) > len(backward_cit):
        print(f"   ‚Ä¢ Innovation Pattern: Creating new knowledge and driving future innovation")
    else:
        print(f"   ‚Ä¢ Innovation Pattern: Balanced knowledge creation and utilization")
else:
    print("‚ö†Ô∏è Citation analysis data not available - Check pipeline execution")
```
